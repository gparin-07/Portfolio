```python
import os
import time
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision.utils import save_image
import matplotlib.pyplot as plt
from PIL import Image
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer
import nltk
from nltk.corpus import stopwords
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Initialize NLTK
nltk.download('punkt')
nltk.download('stopwords')

# Configuration
LR = 0.002
EPOCHS = 5
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class ActivityTracker:
    def __init__(self):
        self.session_log = []
        self.training_stats = []
        self.generated_images = []

    def log_action(self, action_type, details):
        self.session_log.append({
            'type': action_type,
            'details': details,
            'timestamp': time.strftime("%Y-%m-%d %H:%M:%S")
        })

    def log_training(self, epoch_losses, operation):
        self.training_stats.append({
            'operation': operation,
            'epoch_losses': epoch_losses
        })

    def log_image(self, image_path):
        self.generated_images.append(image_path)

    def generate_report(self):
        print("\n===== SESSION SUMMARY =====")

        # 1. Print all actions in order
        print("\nTIMELINE OF ACTIONS:")
        for idx, log in enumerate(self.session_log, 1):
            print(f"{idx}. [{log['timestamp']}] {log['type']}: {log['details']}")

        # 2. Show training statistics
        print("\nTRAINING STATISTICS:")
        for stat in self.training_stats:
            print(f"\nAfter {stat['operation']}:")
            for epoch, loss in enumerate(stat['epoch_losses'], 1):
                print(f"  Epoch {epoch}: Loss = {loss:.4f}")

            # Plot training curve for this operation
            plt.figure(figsize=(6,3))
            plt.plot(stat['epoch_losses'], marker='o')
            plt.title(f'Training after {stat["operation"]}')
            plt.xlabel('Epoch')
            plt.ylabel('Loss')
            plt.grid(True)
            plt.show()

        # 3. Show generated images
        if self.generated_images:
            print("\nGENERATED IMAGES REVIEW:")
            for img_path in self.generated_images:
                try:
                    img = Image.open(img_path)
                    plt.imshow(img)
                    plt.axis('off')
                    plt.title(os.path.basename(img_path))
                    plt.show()
                except Exception as e:
                    print(f"Couldn't load image {img_path}: {str(e)}")

        # 4. Show heatmap of activity frequency
        if self.session_log:
            actions = [log['type'] for log in self.session_log]
            unique_actions = list(set(actions))
            counts = [actions.count(a) for a in unique_actions]

            plt.figure(figsize=(8,4))
            sns.heatmap(np.array([counts]),
                        annot=True,
                        xticklabels=unique_actions,
                        cmap='YlGnBu')
            plt.title('Activity Frequency Heatmap')
            plt.tight_layout()
            plt.show()

class ImageGenerator(nn.Module):
    def __init__(self, latent_dim=100, img_size=64):
        super().__init__()
        self.img_size = img_size

        self.model = nn.Sequential(
            nn.Linear(latent_dim, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, 512),
            nn.LeakyReLU(0.2),
            nn.Linear(512, img_size*img_size*3),
            nn.Tanh()
        )

    def forward(self, z):
        img = self.model(z)
        return img.view(-1, 3, self.img_size, self.img_size)

class SocialMediaAI:
    def __init__(self):
        self.tracker = ActivityTracker()
        self.img_generator = ImageGenerator().to(DEVICE)
        self.optimizer = optim.Adam(self.img_generator.parameters(), lr=LR)
        self.criterion = nn.MSELoss()

        self._init_nlp()
        self._init_chatbot()

    def _init_nlp(self):
        self.tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
        self.sentiment_model = AutoModelForSequenceClassification.from_pretrained(
            "distilbert-base-uncased-finetuned-sst-2-english"
        ).to(DEVICE)
        self.stop_words = set(stopwords.words('english'))
        self.tracker.log_action("System", "NLP components initialized")

    def _init_chatbot(self):
        self.chatbot = ChatBot("SocialAI",
                               storage_adapter='chatterbot.storage.SQLStorageAdapter',
                               database_uri=None,
                               logic_adapters=[
                                   'chatterbot.logic.BestMatch'
                               ])
        trainer = ListTrainer(self.chatbot)

        conversations = [
            ("Hi", "Hello! How can I help you today?"),
            ("How are you?", "I'm functioning optimally, thank you!"),
            ("What do you do?", "I can generate images, analyze text sentiment, and chat with you!"),
            ("Generate an image", "I'll create a new image for you"),
            ("Analyze my text", "Please provide the text you'd like me to analyze"),
            ("Thank-you", "You're welcome!"),
            ("help", "I can: 1) Generate images, 2) Analyze sentiment, 3) Chat, 4) Do calculations")
        ]

        try:
            for pair in conversations:
                trainer.train([pair])
                self.tracker.log_action("System", "Chatbot initialized with basic conversations")
        except Exception as e:
            print(f"Chatbot training error: {str(e)}")
            self.tracker.log_action("System", "Chatbot basic initialization only")


    def _train(self, operation_name):
        epoch_losses = []

        # Silent training - only collect data
        for epoch in range(EPOCHS):
            self.optimizer.zero_grad()
            z = torch.randn(8, 100).to(DEVICE)
            fake_images = self.img_generator(z)
            real_images = torch.rand_like(fake_images)
            loss = self.criterion(fake_images, real_images)
            loss.backward()
            self.optimizer.step()
            epoch_losses.append(loss.item())

        # Log training statistics but don’t show graphs yet
        self.tracker.log_training(epoch_losses, operation_name)
        return f"Trained for {EPOCHS} epochs on {operation_name}"

    def generate_image(self):
        try:
            os.makedirs("generated_images", exist_ok=True)
            img_path = f"generated_images/image_{int(time.time())}.png"

            with torch.no_grad():
                z = torch.randn(1, 100).to(DEVICE)
                generated = self.img_generator(z)
                save_image(generated, img_path, normalize=True)

            # Show just the generated image (no training graphs)
            img = Image.open(img_path)
            plt.imshow(img)
            plt.axis('off')
            plt.show()

            # Silent training
            train_result = self._train("image generation")
            self.tracker.log_action("Image Generation", f"Saved to {img_path}")
            self.tracker.log_image(img_path)

            return "Image generated successfully!"
        except Exception as e:
            return f"Error: {str(e)}"

    def analyze_sentiment(self, text):
        try:
            inputs = self.tokenizer(text, return_tensors="pt").to(DEVICE)
            with torch.no_grad():
                outputs = self.sentiment_model(**inputs)

            probs = torch.nn.functional.softmax(outputs.logits, dim=1)[0]
            sentiment = "positive" if probs[1] > probs[0] else "negative"

            # Silent training
            train_result = self._train("sentiment analysis")
            self.tracker.log_action("Sentiment Analysis",
                                    f"Text: '{text[:30]}...' | Sentiment: {sentiment}")

            return {
                "sentiment": sentiment,
                "confidence": float(max(probs))
            }
        except Exception as e:
            return {"error": str(e)}

    def chat(self, message):
        try:
            response = str(self.chatbot.get_response(message))

            # Silent training
            self._train("chat interaction")
            self.tracker.log_action("Chat", f"Q: {message[:50]}...")

            return {"response": response}
        except Exception as e:
            return {"error": str(e)}

    def calculate(self, expression):
        try:
            allowed_chars = set("0123456789+-*/(). ")
            if not all(c in allowed_chars for c in expression):
                raise ValueError("Only basic math allowed")

            result = eval(expression, {"__builtins__": None}, {})

            # Silent training
            self._train("calculation")
            self.tracker.log_action("Calculation",
                                    f"Expression: {expression} = {result}")

            return {"result": result}
        except Exception as e:
            return {"error": str(e)}

def main():
    print("=== AI Social Media Assistant ===")
    print("1. Generate Image")
    print("2. Analyze Sentiment")
    print("3. Chat")
    print("4. Calculator")
    print("5. View Session Report")
    print("6. Exit")

    ai = SocialMediaAI()

    while True:
        choice = input("\nChoose an option (1-6): ").strip()

        if choice == "1":
            print(ai.generate_image())

        elif choice == "2":
            text = input("Enter text to analyze: ")
            result = ai.analyze_sentiment(text)
            if "error" in result:
                print(f"Error: {result['error']}")
            else:
                print(f"Sentiment: {result['sentiment'].upper()}")
                print(f"Confidence: {result['confidence']:.2f}")

        elif choice == "3":
            print("Chat Mode (type 'exit' to return)")
            while True:
                msg = input("You: ")
                if msg.lower() == 'exit':
                    break
                response = ai.chat(msg)
                if "error" in response:
                    print(f"Error: {response['error']}")
                else:
                    print(f"AI: {response['response']}")

        elif choice == "4":
            print("Calculator Mode (type 'exit' to return)")
            while True:
                expr = input("Expression: ")
                if expr.lower() == 'exit':
                    break
                result = ai.calculate(expr)
                if "error" in result:
                    print(result['error'])
                else:
                    print(f"Result: {result['result']}")

        elif choice == "5":
            ai.tracker.generate_report()

        elif choice == "6":
            print("\nFinal Summary:")
            ai.tracker.generate_report()
            print("\nGoodbye!")
            break

        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
